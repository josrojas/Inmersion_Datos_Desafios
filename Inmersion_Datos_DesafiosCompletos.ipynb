```python
inmuebles["Precio/m2"] = inmuebles["Precio_Millon"] / inmuebles["Area"]
inmuebles.head(3)
```

Para visualizar el resultado de la operación

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

Agrupar los datos de cada barrio para luego obtener el valor exacto de m2 por barrio.

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

Valor real de m2 por barrio

```python
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
```

Diccionario de datos

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head(5)
```

Muestra una columna con los valores x m2 de los barrios

```python
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True)
datos_barrio
```

Muestra la tabla sin indice de barrio y reseteandola a como estaba antes pero con las nuevas columnas y datos de valor_m2_barrio

```python
datos_barrio.query('Barrio in @top_barrios')
```

Mostrar los 10 barrios con mayor valor por m2

```python
precio_m2_barrio = inmuebles.groupby("Barrio")["Valor_m2_Barrio"].mean().sort_values()

plt.figure(figsize=(15,6))

grafica = sns.barplot(data=inmuebles, y='Valor_m2_Barrio', x='Barrio')
grafica.set_title('Valor m2 por barrio', fontsize ='20')

plt.ylabel('Valor de m2', fontsize ='14')
plt.xlabel('Barrio', fontsize ='16')

plt.xticks(rotation=90)

plt.savefig('valor_inmuebles.png',format='png')
plt.show()
```

Se muestra el valor de m2 por barrio pero para hacer más entendible la tabla y tener un visión más clara, se reducirá el número a 10 en el siguiente código.

```python
plt.figure(figsize=(10,8))

ax = sns.barplot(x="Barrio", y="Valor_m2_Barrio", data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)

ax.set_title('Precio m2 por barrio', fontsize ='20')

plt.ylabel('Valor de m2', fontsize ='14')
plt.xlabel('Barrio', fontsize ='16')

plt.savefig('valor_inmuebles.png',format='png')
plt.show()
```

Ahora se tiene una visión más clara y entendible del valor de m2 por cada barrio gracias al uso de data query de top_barrios.
